```python
import pandas as pd
import numpy as np
```

```python
# Création des différents DataFrames
tournaments = pd.read_csv('Sources/atp-world-tour-tennis-data-master/csv/1_tournaments/tournaments_2000_2022.csv', index_col=0).reset_index(drop=True)
match_scores = pd.read_csv('Sources/atp-world-tour-tennis-data-master/csv/2_match_scores/match_scores_2000_2022.csv', index_col=0).reset_index(drop=True)
match_stats = pd.read_csv('Sources/atp-world-tour-tennis-data-master/csv/3_match_stats/match_stats_2000_2022.csv', index_col=0).reset_index(drop=True)
odds_data = pd.read_csv('Sources/tennis-data.co.uk-with-odds/odds_data_2000_2022.csv', index_col=0).reset_index(drop=True)
```

```python
match_scores_tournaments = match_scores.merge(tournaments, on='tourney_year_id', how='inner')
```

```python
datasets_without_odds = match_scores_tournaments.merge(match_stats, on='match_id', how='left')
```

```python
datasets_without_odds['match_score_tiebreaks']
```

```python
set_1 = []
set_2 = []
set_3 = []
set_4 = []
set_5 = []

for value in datasets_without_odds['match_score_tiebreaks']:
    if len(value.split()) == 1:
        set_1.append(value.split()[0])
        set_2.append(np.nan)
        set_3.append(np.nan)
        set_4.append(np.nan)
        set_5.append(np.nan)
    elif len(value.split()) == 2:
        set_1.append(value.split()[0])
        set_2.append(value.split()[1])
        set_3.append(np.nan)
        set_4.append(np.nan)
        set_5.append(np.nan)
    elif len(value.split()) == 3:
        set_1.append(value.split()[0])
        set_2.append(value.split()[1])
        set_3.append(value.split()[2])
        set_4.append(np.nan)
        set_5.append(np.nan)
    elif len(value.split()) == 4:
        set_1.append(value.split()[0])
        set_2.append(value.split()[1])
        set_3.append(value.split()[2])
        set_4.append(value.split()[3])
        set_5.append(np.nan)
    elif len(value.split()) == 5:
        set_1.append(value.split()[0])
        set_2.append(value.split()[1])
        set_3.append(value.split()[2])
        set_4.append(value.split()[3])
        set_5.append(value.split()[4])
    else:
        set_1.append(np.nan)
        set_2.append(np.nan)
        set_3.append(np.nan)
        set_4.append(np.nan)
        set_5.append(np.nan)
```

```python
score_1 = pd.DataFrame(set_1)
score_1 = score_1.astype('str')
set_1 = list(score_1[0])

score_2 = pd.DataFrame(set_2)
score_2 = score_2.astype('str')
set_2 = list(score_2[0])

score_3 = pd.DataFrame(set_3)
score_3 = score_3.astype('str')
set_3 = list(score_3[0])

score_4 = pd.DataFrame(set_4)
score_4 = score_4.astype('str')
set_4 = list(score_4[0])

score_5 = pd.DataFrame(set_5)
score_5 = score_5.astype('str')
set_5 = list(score_5[0])
```

```python
W1 = []
L1 = []
W2 = []
L2 = []
W3 = []
L3 = []
W4 = []
L4 = []
W5 = []
L5 = []

for value in set_1:
    W1.append(value[0])
    L1.append(value[1])

for value in set_2:
    if value == np.nan:
        W2.append(np.nan)
        L2.append(np.nan)
    else:
        W2.append(value[0])
        L2.append(value[1])
        
for value in set_3:
    if value == np.nan:
        W3.append(np.nan)
        L3.append(np.nan)
    else:
        W3.append(value[0])
        L3.append(value[1])
        
for value in set_4:
    if value == np.nan:
        W4.append(np.nan)
        L4.append(np.nan)
    else:
        W4.append(value[0])
        L4.append(value[1])
        
for value in set_5:
    if value == np.nan:
        W5.append(np.nan)
        L5.append(np.nan)
    else:
        W5.append(value[0])
        L5.append(value[1])
```
